```python
import pandas as pd
```

```python
users = pd.read_csv("../dataset/data/users.csv")

display(users.head())

print(users.shape)
```

### Create login_name: user_id dictionary

```python
# collect all login names to a list
all_logins = users.login.values.tolist()
all_ids = users.id.values.tolist()

# create a dictionary to map login names to ids
login2id = dict(zip(all_logins, all_ids))

login2id
```

### Create (user_id, following_list) and (user_id, follower_list) tuple list

```python
all_followings = users.following.values.tolist()
all_followers = users.followers.values.tolist()

# do eavl() to convert string to list
all_followings = [eval(following) for following in all_followings]
all_followers = [eval(follower) for follower in all_followers]

# create (id, following) tuple list
all_followings = [(id, following) for id, following in zip(all_ids, all_followings)]
all_followers = [(id, follower) for id, follower in zip(all_ids, all_followers)]
```

### Check desnity of the graph

```python
FOLLOWING_THRESHOLD = 2
FOLLOWER_THRESHOLD = 2

def get_users_in_network(list, threshold: int):
    unreach_users_count = 0
    for (_, user_list) in list:
        # filter any login_name in followings that is not in all_logins
        users_in_dataset = [user for user in user_list if user in login2id]
        if len(users_in_dataset) < threshold:
            unreach_users_count += 1
    return unreach_users_count

unreach_followings_count = get_users_in_network(all_followings, FOLLOWING_THRESHOLD)
unreach_followers_count = get_users_in_network(all_followers, FOLLOWER_THRESHOLD)

print(f"total # users in network: {users.shape[0]}")
print(f"# users having following count < {FOLLOWING_THRESHOLD}: {unreach_followings_count}, propotion: {unreach_followings_count / users.shape[0]:.2f}")
print(f"# users having follower count < {FOLLOWER_THRESHOLD}: {unreach_followers_count}, propotion: {unreach_followers_count / users.shape[0]:.2f}")
```

### Save the user_following_dict and user_follower_dict to pickle file

```python
def filter_and_transform_user_list(list):
    user_connection_dict = {}
    for (user_id, user_list) in list:
        users_in_dataset = [user for user in user_list if user in login2id]
        transformed_ids = [login2id[user] for user in users_in_dataset]
        user_connection_dict[user_id] = transformed_ids
    return user_connection_dict
```

```python
filtered_followings_dict = filter_and_transform_user_list(all_followings)
filtered_followers_dict = filter_and_transform_user_list(all_followers)
```

```python
import pickle

with open("../dataset/data/users_followings_dict.pkl", "wb") as f:
    pickle.dump(filtered_followings_dict, f)

with open("../dataset/data/users_followers_dict.pkl", "wb") as f:
    pickle.dump(filtered_followers_dict, f)
```
