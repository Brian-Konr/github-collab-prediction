```python
import pandas as pd
```

```python
users = pd.read_csv("../dataset/data/users.csv")

display(users.head())

print(users.shape)
```

```python
# make bio column string
users["bio"] = users["bio"].astype(str)
```

### Use Sentence Transformer to encode bio sentences into vectors

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
bio_embeddings = {}
for i, user in users.iterrows():
    bio_embeddings[user["id"]] = model.encode(user["bio"])

# save bio_embeddings to pickle file
import pickle
with open("../dataset/data/bio_embeddings.pkl", "wb") as f:
    pickle.dump(bio_embeddings, f)
```
